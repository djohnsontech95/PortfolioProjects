```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (10,6)

# Read in the data 

df = pd.read_csv(r'C:\Users\Dylan\Documents\Personal\Projects\Python projects\Data Analysis\movie-industry\movie-data.csv')
```

```python
# Quick look at data

df.head()
```

```python
# Check for any missing data

for col in df.columns: 
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data frame before dropna()

df.head()
```

```python
# Drop all rows with missing data and reset indexing

new_df = df.dropna().reset_index(drop=True)
```

```python
# Clean data frame

new_df.head()
```

```python
# Look at data types of columns

new_df.dtypes
```

```python
# Change data type of columns

new_df['votes'] = new_df['votes'].astype('int64')

new_df['budget'] = new_df['budget'].astype('int64')

new_df['gross'] = new_df['gross'].astype('int64')

new_df['runtime'] = new_df['runtime'].astype('int64')
```

```python
# Create correct year column

new_df['yearcorrect'] = new_df['released'].astype(str).str.split().str[2]
```

```python
new_df.head()
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=new_df['budget'], y=new_df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget for Film')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
# Sort values by budget to see data easier on graph

new_df.sort_values(by=['budget'], inplace=False, ascending=False)
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=new_df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
# Quick look at correlation

correlation_matrix = new_df.corr(method='pearson')
```

```python
# Correlation Matrix heat map

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Change object columns to category type for counting

df_numerized = new_df

for col_name in df_numerized.columns: 
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()

new_df.sort_values(by=['budget'], inplace=False, ascending=False)
```

```python
# For easy comparison to numerical values above to ensure correct data

new_df.sort_values(by=['budget'], inplace=False, ascending=False)
```

```python
# Correlation Matrix heat map

correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Correlation heat map unstacked for easier visualizing and comparison

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# Sorted correlations

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Shows only correlations higher than 0.5

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
