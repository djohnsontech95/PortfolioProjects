```python
### Counts the number of goals per team
goals_per_team = df_goals['team_name'].value_counts()

### Top 10 goals scored per team
top_10_goals = goals_per_team.nlargest(10)

### Bottom 10 on gaols scored
lowest_10_goals = goals_per_team.nsmallest(10)

### Print results
print("Top 10 teams with the most goals scored:")
print(top_10_goals)
print("\nBottom 10 teams with the least amount of goals scored:")
print(lowest_10_goals)
```

```python
### Create a new column with the year of each match
df_goals['match_year'] = df_goals['match_date'].astype(str).str[:4]
```

```python
### Group by match_year and team_name to organize by year and count the number of team occurances
### Creates a new column called yearly_goals to hold all the new data
yearly_goals_per_team = df_goals.groupby(['match_year', 'team_name']).size().reset_index(name='goals_in_year')

print(yearly_goals_per_team)
```

```python
### Pivot the data to create a matrix with countries as rows, years as columns, and total goals as values
yearly_goals_heatmap = yearly_goals_per_team.pivot('team_name', 'match_year', 'goals_in_year')

## Create the heatmap
plt.figure(figsize=(15, 14))
sns.heatmap(yearly_goals_heatmap, cmap='YlGnBu', annot=True, fmt='g', linewidths=.5)
plt.title('Total Goals Scored by Each Team Over the Years')
plt.xticks(rotation=45)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Teams', fontsize=14)
plt.show()
```

```python
### Setup data
team_data = yearly_goals_per_team.pivot('match_year', 'team_name', 'goals_in_year')

### Calculate the yearly percentage change for each team
percentage_change = team_data.pct_change() * 100

### Identify teams with the most significant positive changes
most_improved_teams = percentage_change.max().nlargest(5)

### Identify teams with the most significant negative changes
most_declined_teams = percentage_change.min().nsmallest(5)

print("Most Improved Teams:")
print(most_improved_teams)
print("\nMost Declined Teams:")
print(most_declined_teams)
```

```python
### Calculate the total goals scored per year
total_goals_per_year = yearly_goals_per_team.groupby('match_year')['goals_in_year'].sum()

### Merge the total goals information back into the yearly_goals_per_team DataFrame
yearly_goals_per_team = pd.merge(yearly_goals_per_team, total_goals_per_year.rename('total_goals'), on='match_year', how='left')

### Calculate the probability that a team scores in a given year
yearly_goals_per_team['probability'] = (yearly_goals_per_team['goals_in_year'] / yearly_goals_per_team['total_goals']) * 100

print(yearly_goals_per_team)
```

```python
### Find the top 5 highest probabilities across all rows
top_5_probabilities = yearly_goals_per_team['probability'].nlargest(5)

### Find the corresponding team and year for the top 5 probabilities
top_5_teams_years = yearly_goals_per_team.loc[yearly_goals_per_team['probability'].isin(top_5_probabilities)].reset_index(drop=True)

### Display the result
print(top_5_teams_years)
```

```python
### Count the occurrences of each unique tournament_id
tournament_match_counts = df_team_appearances['tournament_id'].value_counts()

### Divide by 2 to get the actual number of matches
actual_number_of_matches = tournament_match_counts / 2

### Sorts the index in DESC order
actual_number_of_matches = actual_number_of_matches.sort_index()
```

```python
### Reset the index of actual_number_of_matches to create a DataFrame
number_of_matches_df = actual_number_of_matches.reset_index()
### Create column names for entries
number_of_matches_df.columns = ['tournament_id', 'total_matches']
### Create new column for match_year
number_of_matches_df['match_year'] = number_of_matches_df['tournament_id'].str[-4:]

### Merge DataFrames on the common column 'match_year'
merged_df = pd.merge(yearly_goals_per_team, number_of_matches_df, on='match_year')

### Add a new column 'total_matches' to yearly_goals_per_team
yearly_goals_per_team['total_matches'] = merged_df['total_matches']

print(yearly_goals_per_team)
```

```python
### Calculate the average goals per match for each unique 'match_year'
yearly_goals_per_team['average_goals_per_match'] = yearly_goals_per_team['total_goals'] / yearly_goals_per_team['total_matches']

### Round the 'average_goals_per_match' column to two decimal places
yearly_goals_per_team['average_goals_per_match'] = yearly_goals_per_team['average_goals_per_match'].round(2)

print(yearly_goals_per_team)
```

```python
### Top 5 years with the highest amount of goals scored on average
top_5_years = yearly_goals_per_team.groupby('match_year')['average_goals_per_match'].max().nlargest(5)

print("Top 5 years with the highest amount of goals scored on average:")
print(top_5_years)
```

```python
### Count the number of unique tournament IDs 
world_cups_count = df_team_appearances['tournament_id'].nunique()

### Drop duplicate rows based on 'match_year' (only to figure out average across all World Cups)
unique_goals_per_year = yearly_goals_per_team.drop_duplicates(subset='match_year')

### Calculate the overall average goals across all World Cups
overall_average_goals = unique_goals_per_year['average_goals_per_match'].mean().round(2)

### Display overall average goals result
print("Overall average goals scored in any given World Cup:")
print(overall_average_goals)
```

```python
### Create a line plot to visualize the change in average goals per World Cup over the years
sns.lineplot(x=yearly_goals_per_team['match_year'], y=yearly_goals_per_team['average_goals_per_match'], marker='o', color='skyblue')

### Set plot title and labels
plt.title('Change in Average Goals per World Cup Over the Years', fontsize=16)
plt.xlabel('Match Year', fontsize=14)
plt.ylabel('Average Goals per World Cup', fontsize=14)

### Rotate x-axis labels for better readability
plt.xticks(rotation=45)

## Change in average goals per world cup over the years
plt.show()
```

```python
### Filter the data for the year 1954
year_1954_data = yearly_goals_per_team[yearly_goals_per_team['match_year'] == '1954']


### Grab team_name and goals_in_year from this data
result = year_1954_data[['team_name', 'goals_in_year']]

#### print(result['goals_in_year'].max())

## Create visual to analyze data
result.plot(x='team_name', y='goals_in_year', kind='bar', legend=False, color='skyblue')
plt.title('Team Goals in 1954 World Cup')
plt.xlabel('Teams')
plt.ylabel('Number of Goals')
plt.xticks(rotation=45, ha='right')
plt.show()
```
